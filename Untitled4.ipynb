**6.Low level image processing: image in, image out**

Low level processing improves visual quality without extracting semantic meaning, as described in the lecture pipeline  

**Exercise 6.1: Image enhancement by contrast stretching**

```python
min_val = gray.min()
max_val = gray.max()

enhanced = (gray - min_val) / (max_val - min_val) * 255

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Original")

plt.subplot(1,2,2)
plt.imshow(enhanced, cmap='gray')

plt.title("Contrast Enhanced")
plt.show()
```

**Questions**

2. Is this enhancement subjective or objective? objective

**7.Noise and image restoration**

Noise removal is a core example of low level processing.

**Exercise 7.1: Add artificial noise**

```python
noise = np.random.normal(0, 20, gray.shape) #0--> mean , 20--> std
noisy = gray + noise

plt.imshow(noisy, cmap='gray')
plt.title("Noisy Image")
plt.show()
```

**Exercise 7.2: Simple smoothing filter**

```python
def mean_filter(img, k=3):
  pad = k // 2
  padded = np.pad(img, pad, mode='edge')
  output = np.zeros_like(img)
  for i in range(img.shape[0]):
    for j in range(img.shape[1]):
      output[i, j] = padded[i:i+k, j:j+k].mean()
  return output

denoised = mean_filter(noisy)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(noisy, cmap='gray')
plt.title("Noisy")

plt.subplot(1,2,2)
plt.imshow(denoised, cmap='gray')
plt.title("Denoised")
plt.show()
```

**9. Mini project: connecting theory to practice**

**Task**

Using only NumPy and Matplotlib:
1. Create a synthetic image containing at least two objects
2. Add noise to the image
3. Apply enhancement and denoising
4. Segment the objects using a threshold


**1. Create a synthetic image containing at least two objects**

```python
image = np.zeros((256,256))

#obj1 (circle)
x,y =np.meshgrid(np.arange(256),np.arange(256))
circle = (x-60)**2+(y-60)**2<50**2
image[circle]=255

# obj2 (rectangle)
image[120:200,120:250]=255

plt.imshow(image, cmap='gray',vmin=0,vmax=255)
plt.title("image")
plt.colorbar()
plt.show()
```

**2. Add noise to the image**

```python
noise=np.random.normal(0,30,image.shape) # 0 --> mean, 30 --> std
image_noisy=image+noise
image_noisy= np.clip(image_noisy,0,255)# to keep it in the range

plt.imshow(image_noisy, cmap='gray',vmin=0,vmax=255)
```

**3. Apply enhancement and denoising**

```python
enhanced = (image_noisy - image_noisy.min()) / (image_noisy.max() - image_noisy.min()) * 255
def mean_filter(enhanced, k=3):
  pad = k // 2
  padded = np.pad(enhanced, pad, mode='edge')
  output = np.zeros_like(enhanced)
  for i in range(enhanced.shape[0]):
    for j in range(enhanced.shape[1]):
      output[i, j] = padded[i:i+k, j:j+k].mean()
  return output

denoised = mean_filter(enhanced)

plt.figure(figsize=(10,4))
plt.subplot(1,3,1)
plt.imshow(image_noisy, cmap='gray', vmin=0, vmax=255)
plt.title("Noisy")

plt.subplot(1,3,2)
plt.imshow(enhanced, cmap='gray', vmin=0, vmax=255)
plt.title("Enhanced")

plt.subplot(1,3,3)
plt.imshow(denoised, cmap='gray', vmin=0, vmax=255)
plt.title("Denoised")
plt.show()
```

**4. Segment the objects using a threshold**

```python
threshold=120 # for each pixl
binary = denoised > threshold # pixels > threshold  1 (white) # pixels <= threshold  0 (black)

plt.imshow(binary, cmap='gray')
plt.title("Binary Segmentation")
plt.show()
```
